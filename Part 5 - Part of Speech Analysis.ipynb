```python
def tokenizer(post_0):
    noun = []
    nouncount = 0
    verb = []
    verbct = 0
    verb = []
    verbct = 0
    prepositions = []
    prepct =0

    for x in post_0:
        if x[1] == 'NN' or x[1]=='NNS':
            noun.append(x[0])
            nouncount = nouncount + 1
    
        if x[1] == 'VBP' or x[1] =='VBG' or x[1] == 'VBZ' or x[1] =='VBD' or x[1]=='VB' or x[1]=='VBN':
            verb.append(x[0])
            verbct = verbct + 1
        if x[1] == 'PRP':
            prepositions.append(x[0])
            prepct = prepct + 1
    
    return (nouncount,verbct,prepct)
```
