```python
cv = pd.read_csv('./countvectorized posts/500feat_cv.csv')
```

```python
X = cv.drop(cv[['Unnamed: 0','target','i_e','s_n','t_f','j_p']], axis=1)
y_i_e = cv.i_e
y_s_n = cv.s_n
y_t_f = cv.t_f
y_j_p = cv.j_p
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score, train_test_split
```

```python
kfold = StratifiedKFold()
logreg=LogisticRegression()
```

```python
X_train, X_test, y_e_train, y_e_test = train_test_split(X,y_i_e)
```

```python
logreg.fit(X_train,y_e_train)
```

```python
cross_val_score(logreg, X_test, y_e_test, cv=kfold, scoring='accuracy').mean()
```

```python
logreg.coef_
```

```python
trans = logreg.coef_.T
```

```python
df = pd.DataFrame(zip(trans, X_train.columns))
df.columns = ['coef','word']
```

```python
df.coef = [x[0] for x in trans]
```

```python
df.head()
```

```python
df['vals'] = abs(df.coef)
```

```python
df.head()
```

```python
#Words that represent EXTROVERSION
df.sort_values('vals',ascending=False)
```

```python
df = df[['word','vals','coef']]
```

```python
#words that represent EXTROVERSION
df.sort_values('coef', ascending=False).head(25)
```

```python
#most predictive of EXTRAVERSION
e_predictors = df.sort_values('coef', ascending=False).head(25)
e_predictors.to_csv('./Word Lists/e_predict.csv')
```
