```python
import pandas as pd
import numpy as np
```

```python
cv = pd.read_csv('./500feat_cv.csv')
```

```python
cv.head()
```

```python
X = cv.drop(cv[['target','Unnamed: 0','i_e','s_n','t_f','j_p']], axis=1)
y_i_e = cv.i_e
y_s_n = cv.s_n
y_t_f = cv.t_f
y_j_p = cv.j_p
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold, cross_val_score, train_test_split
```

```python
X_train, X_test, y_f_train, y_f_test = train_test_split(X,y_t_f, random_state=42)
```

```python
logregf = LogisticRegression()
```

```python
logregf.fit(X_train,y_f_train)
```

```python
kfold = StratifiedKFold()
cross_val_score(logregf, X_test, y_f_test, cv=kfold, scoring='accuracy').mean()
```

```python
df = pd.read_csv('./1000feat_cv.csv')
```

```python
X = df.drop(df[['target','i_e','s_n','t_f','j_p']], axis=1)
y_i_e = df.i_e
y_s_n = df.s_n
y_t_f = df.t_f
y_j_p = df.j_p
```

```python
X_train, X_test, y_f_train, y_f_test = train_test_split(X,y_t_f, random_state=42)
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(X_train,y_f_train)
```

```python
kfold = StratifiedKFold()
cross_val_score(logreg, X_test, y_f_test, cv=kfold, scoring='accuracy').mean()
```

```python
#remember that baseline is 54, score is 76! :)
```

```python
df = pd.DataFrame(zip(logreg.coef_.T, X_train.columns))
df.columns = ['coef','word']
```

```python
df.coef = [x[0] for x in df.coef]
```

```python
df['abs'] = abs(df.coef)
```

```python
df = df[['word','abs','coef']]
```

```python
#most important words for predicting T or F
df.sort_values('abs', ascending=False).head(25)
```

```python
torf_predictors = df.sort_values('abs', ascending=False).head(25)
torf_predictors.to_csv('./Word Lists/torf_predictors.csv')
```
