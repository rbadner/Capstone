```python
import pandas as pd
import numpy as np
```

```python
cv = pd.read_csv('./countvectorized posts/500feat_cv.csv')
```

```python
cv.head()
```

```python
X = cv.drop(cv[['Unnamed: 0','target','i_e','s_n','t_f','j_p']], axis=1)
y_i_e = cv.i_e
y_s_n = cv.s_n
y_t_f = cv.t_f
y_j_p = cv.j_p
```

```python
X.columns
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score, train_test_split
```

```python
X_train, X_test, y_n_train, y_n_test = train_test_split(X,y_s_n)
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(X_train,y_n_train)
```

```python
kfold = StratifiedKFold()
cross_val_score(logreg, X_test, y_n_test, cv=kfold, scoring='accuracy').mean()
```

```python
df = pd.DataFrame(zip(logreg.coef_.T, X_train.columns))
df.columns = ['coef','word']
```

```python
df.coef = [x[0] for x in df.coef]
```

```python
df['abs'] = abs(df.coef)
```

```python
df = df[['word','abs','coef']]
```

```python
#most important words for predicting S or N
df.sort_values('abs', ascending=False).head(25)
```

```python
#highest predictors of iNtuition
df.sort_values('coef', ascending=False).head(25)
```

```python
n_predictors = df.sort_values('coef', ascending=False).head(25)
n_predictors.to_csv('./Word Lists/n_predict.csv')
```

```python
#highest predictors of SENSING
s_predictors = df.sort_values('coef', ascending=True).head(25)
```
